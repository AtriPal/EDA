## Q.12-> What is the distribution of left-footed vs. right-footed players?

```python
foot_count = df['Foot'].value_counts()
foot_count
```

```python
foot_count.values
```

```python
ex=[0.1,0.05,0.0,0.0]
plt.pie(foot_count.values, labels= foot_count.index, autopct='%1.1f%%', explode=ex)
```

## Q.13-> Does foot preference affect goals scored?

```python
val1 = df.groupby('Foot')['Goals'].mean().values
```

```python
ind1= df.groupby('Foot')['Goals'].mean().index
```

```python
plt.pie(val1, labels=ind1, autopct='%1.1f%%')
```

```python
fig, axes = plt.subplots(1,2, figsize=(12,6))

axes[0].pie(val1, labels=ind1, autopct='%1.1f%%', colors=['lightyellow', 'lightblue','grey','lightgreen'])
axes[0].set_title("Avg Goals by Foot preferance")

sns.boxplot(data=df, x='Goals', y='Foot', ax=axes[1], palette=['lightgreen', 'grey','lightblue','lightyellow'])
axes[1].set_title('Goal Distribution by Foot Preference')
axes[1].set_xlabel('Foot')
axes[1].set_ylabel('Goals')

plt.tight_layout()
plt.show()
```
